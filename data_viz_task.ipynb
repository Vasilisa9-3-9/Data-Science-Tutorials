```python
# Import libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Load data
car_data = pd.read_csv('Cars93.csv', index_col = 0, keep_default_na = False) 
```

```python
car_data.columns
```

### 1. Boxplot

```python
# Boxplot for revs per mile
manufacturer = ["Audi","Suzuki","Toyota","Hyundai"]
manufacturer_data = car_data[car_data["Manufacturer"].isin(manufacturer)]
sns.boxplot( x = manufacturer_data["Manufacturer"], y=manufacturer_data["Rev.per.mile"] )
```

```python
manufacturer_data
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**
Suzuki 


### 2. Histogram

```python
# Histogram of MPG
#sns.displot( a = car_data["MPG.city"], hist=True, kde=False, rug=False )
sns.histplot(data = car_data, x="MPG.city", color="skyblue", label="MPG.city", kde=True)
sns.histplot(data=car_data, x="MPG.highway", color="red", label="MPG.highway", kde=True)
plt.legend() 
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**
city 

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
sns.lineplot(x = car_data["Wheelbase"], y = car_data["Turn.circle"])
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?



**Lineplot answer:**  Overall, a positive correlation can be noticed between two variables. When the wheelbase gets larger the turning circle also increases. 


### 4. Barplot

```python
# Barplot for mean horsepower
#mean_horseposer = car_data[car_data["Horsepower"].mean()]
sns.barplot( x = car_data["Type"], y = car_data["Horsepower"],  data = car_data, estimator = sum, ci=None, color='#69b3a2')
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:** No, according to the barplot, the midsize car type has the highest horsepower
